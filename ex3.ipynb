Of course entangled states can take many different forms. For instance consider a $3$ qubit state that has an even number of ones
$$|\text{Even}\rangle = c_{00}|000\rangle + c_{01}|101\rangle + c_{10}|110\rangle + c_{11}|011\rangle.$$

If we ignore the first qubit, the state in the other qubits could be non entangled depending on the values of $c_{ij}$. In particular, if we set $c_{ij} = \frac{1}{2}$ we would have an equal superposition state in the last two qubits

$$|++\rangle = \frac{1}{2} \left(|00\rangle+|01\rangle+|10\rangle+|11\rangle\right),$$

that is, a product state $|++\rangle = \left(\frac{1}{\sqrt{2}}\left(|0\rangle+|1\rangle\right)\right)\otimes \left(\frac{1}{\sqrt{2}}\left(|0\rangle+|1\rangle\right)\right)$. But if we add a first qubit that sets the parity,

$$|\text{Even}\rangle_{++} = \frac{1}{2}\left(|000\rangle + |101\rangle + |110\rangle + |011\rangle\right),$$

this product state structure is lost. Additionally, if any of the qubits were to be measured, it is guaranteed that the other pair of qubits is in the state $|\phi^+\rangle$ if $0$ is measured or $|\psi^+\rangle$ if $1$ is measured.

Even if this might seem very trivial, it can be important for real physical systems. Two good examples of the usefulness of this approach are that ground states of some physical systems can have a fixed parity or that this is taken into account in order to implement time evolution operators $e^{-iHt}$ (Nielsen and Chuang section 4.7.3)

# CHSH inequality

We have been discussing some aspects of how entanglement works and how counterintuitive it can really be. As we saw during the lecture, during the early days of Quantum Mechanics, a significative part of the community thought that it could not be correct. In particular, what was particularly controversial was how measurements were supposed to work. If superpositions were not to be something real, but rather just an artefact created by our lack of understanding, there would not be any problem at all. What was proposed as an alternative was that there could be some underlying theory that could explain what the result of a given measurement would be in a deterministic manner. This type of theories were called *hidden variable theories*.

For this to work, *local realism* should hold. This means that, every system should have all its measurable quantities perfectly locally determined. In particular, entangled multipartite systems that were far away could not affect one another after a measurement, everything should be determined beforehand.

In 1963, John Bell discovered that one could test experimentally wether or not local realism holds true.

## Measurements in different basis.

In quantum computing we tend to write states in the computational basis, that is, for a $1$ qubit system,

$$|\psi\rangle = \sqrt{p_0}|0\rangle + \sqrt{1-p_0} |1\rangle.$$

As you saw in the previous exercise, when we measure we say that we get $0$ with a probability $p_0$ and $1$ with a probability $1-p_0$. These measurements are being done along the Z axis in the Bloch sphere, i.e, we are getting the projection of the state along that axis. In a more mathematical sense, this is because $|0>$ and $|1\rangle$ are eigenstates of $Z$ with eigenvalues $+1$ and $-1$.

In this way, $\langle Z \rangle = \langle \psi | Z | \psi \rangle= +1\times p_0 +(-1)\times(1-p_0)$.

What would happen if we wanted to measure along the X axis? First, we have to notice that the eigenvalues of the operator $X$ are

$$|\pm\rangle = \frac{1}{\sqrt{2}}\left(|0\rangle \pm |1\rangle\right) \quad \text{as} \quad X|\pm\rangle = \pm1|\pm\rangle.$$

If we are able to map X eigenstates into Z eigenstates, $|+\rangle \rightarrow |0\rangle$ and $|-\rangle \rightarrow |1\rangle$ we could measure $X$ by first doing that transformation and then measuring Z.

Now that we are able to measure both in Z and X basis, we can make a function that computes the expectation value given a number of measurements `nshots` for a 1 qubit circuit.

```python
def expectation_value_single_qubit(qc, nshots=8092):
    # Run the given circuit that is already measuring the selected basis
    backend_shots = Aer.get_backend('qasm_simulator')
    counts = execute(qc, backend_shots, shots=nshots).result().get_counts() #Counts is a dictionary that contains the number of zeros and the number of ones measured
    
    # This is just to make things easier in your solution
    if '1' not in counts.keys():
        counts['1'] = 0
    if '0' not in counts.keys():
        counts['0'] = 0
        
    # Get the number of times that 0 and 1 were measured
    n_zeros = counts['0']
    n_ones  = counts['1']
    
    # Compute the probabilities
    p_0 = n_zeros/nshots
    p_1 = n_ones/nshots #OR 1-p_0

    expectation_value = 1*p_0+(-1)*p_1
    return expectation_value
```

## Computing CHSH

Imagine now Alice and Bob share each of the components of a bipartite system, separate from each other and both of them simultaneusly measure its component in one basis, chosen randomly from two possibilities. Alice's basis will be called A and a, Bob's basis will be called B and b. What would happen if we measured the expectation value of

$$\langle CHSH\rangle = \langle AB\rangle - \langle Ab\rangle + \langle aB\rangle + \langle ab\rangle ?$$

If local realism holds true, Alice's measurement should not influence Bob's. Additionally the properties that we measure should be defined before the experiment is done. As each measurement can only yield the values $\pm1$, both $(B-b)$ and $(B+b)$ can only be $0$ or $\pm2$. Consequently, $A(B-b)+a(B+b)$ can only be $+2$ or $-2$, so $|\langle CHSH\rangle| \leq 2$.

Let's see what happens if we replicate this experiment quantumly, using the state $|\phi^+\rangle$. We will assume that Bob's basis are $B=Z$ and $b=X$ while Alice's basis are $A=\frac{1}{\sqrt{2}}\left(Z-X\right)$ and $a=\frac{1}{\sqrt{2}}\left(Z+X\right)$. This basically means that Alice's basis are rotated an angle $\pi/2$ with respect of Bob's.

Notice that we can write, for instance $\langle \phi^+|AB|\phi^+\rangle = \langle \phi^+|\frac{1}{\sqrt{2}}\left(Z-X\right)Z|\phi^+\rangle = \frac{1}{\sqrt{2}}\left(\langle \phi^+|ZZ|\phi^+\rangle-\langle \phi^+|XZ|\phi^+\rangle\right)$ so we can decompose everything into measuring $ZZ$, $ZX$, $XZ$ and $XX$.
